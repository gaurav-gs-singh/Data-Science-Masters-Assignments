### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

Different plots used under seaborn are:
1. Distplot : Dist plot gives us the histogram of the selected continuous variable. It is an example of a univariate analysis. We can change the number of bins i.e. number of vertical bars in a histogram.
2. Joint Plot: It is the combination of the distplot of two variables. It is an example of bivariate analysis. We additionally obtain a scatter plot between the variable to reflecting their linear relationship.
3. Pair Plot: It takes all the numerical attributes of the data and plot pairwise scatter plot for two different variables and histograms from the same variables.
4. Bar Plot: It is an example of bivariate analysis. On the x-axis, we have a categorical variable and on the y-axis, we have a continuous variable.
5. Box Plot: It is a 5 point summary plot. It gives the information about the maximum, minimum, mean, first quartile, and third quartile of a continuous variable. Also, it equips us with knowledge of outliers. We can plot this for a single continuous variable or can analyze different categorical variables based on a continuous variable.

### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset('fmri')
```

```python
fmri.head()
```

```python
sns.lineplot(x = fmri.timepoint, y = fmri.signal, data = fmri, hue = 'event', style = 'region')
```

### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.age)
```

```python
sns.boxplot(x = titanic.pclass, y = titanic.fare)
```
