```python
import plotly.graph_objects as go
import seaborn as sns
```

**Q1. Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.**

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x = titanic.age, y = titanic.fare, mode = 'markers'))
```

**Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.**

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y= tips['total_bill'], x= tips['day']))
```

**Q3. Using the tips dataset in the Plotly library, Plot a histogram for x="sex" and y="total_bill" column in the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day" column with the color parameter.**

```python
import plotly.express as px
tips= px.data.tips()
```

```python
tips.head()
```

```python
fig = px.histogram(tips, x = 'sex', y = 'total_bill', pattern_shape='smoker', color = 'day')
fig.show()
```

**Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for the color parameter.
Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the dimensions parameter.**

```python
import plotly.express as px
```

```python
iris = px.data.iris()
iris.head()
```

```python
fig = px.scatter_matrix(iris,dimensions = ['sepal_length',	'sepal_width',	'petal_length','petal_width'], color = 'species')
fig.show()
```

**Q5. What is Distplot? Using Plotly express, plot a distplot.**

Distplots
The distplot figure factory displays a combination of statistical representations of numerical data, such as histogram, kernel density estimation or normal curve, and rug plot.



```python
import numpy as np
import plotly.figure_factory as ff
x = np.random.randn(1000)
hist_data = [x]
group_labels = ['distplot']
fig = ff.create_distplot(hist_data,group_labels)
```

```python
fig.show()
```
